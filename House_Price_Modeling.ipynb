```python
# House Price Modeling Pipeline
# 1) Setup & imports
import os
import json
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, KFold, cross_val_score, GridSearchCV, LeaveOneOut, TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.base import clone
import joblib

# 2) Utility functions (metrics, plotting, VIF)

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def mape(y_true, y_pred):
    # Avoid division by zero: add small epsilon
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    eps = 1e-9
    return np.mean(np.abs((y_true - y_pred) / (y_true + eps))) * 100

def regression_metrics(y_true, y_pred):
    return {
        'RMSE': float(rmse(y_true, y_pred)),
        'MAE': float(mean_absolute_error(y_true, y_pred)),
        'R2': float(r2_score(y_true, y_pred)),
        'MAPE': float(mape(y_true, y_pred))
    }

# VIF requires statsmodels or manual calc
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(X_df):
    """Calculate VIF for dataframe of numeric features (no intercept column)."""
    X = X_df.copy()
    # Add constant for VIF in some implementations; here statsmodels doesn't need it.
    vif_data = pd.DataFrame()
    vif_data['feature'] = X.columns
    vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

# Residual plot
def plot_residuals(y_true, y_pred, title='Residuals'):
    resid = y_true - y_pred
    plt.figure(figsize=(6,4))
    plt.scatter(y_pred, resid, alpha=0.5)
    plt.axhline(0, color='black', linestyle='--')
    plt.xlabel('Predicted')
    plt.ylabel('Residuals')
    plt.title(title)
    plt.show()

# 3) Load data & quick EDA
DATA_PATH = '/content/House Price India.csv'
print(f"Warning: {DATA_PATH} not found. Please place your CSV at this path or edit DATA_PATH.")

# Try to load
try:
    df = pd.read_csv(DATA_PATH)
except Exception as e:
    print('Failed to load CSV:', e)
    df = pd.DataFrame()

if df.shape[0] == 0:

    print('\nNo data loaded - script will show structure. Replace DATA_PATH with your file to run end-to-end.')
else:
    print('Data loaded. Shape:', df.shape)
    display(df.head())

# --- From here on, code assumes df is loaded ---
if df.shape[0] > 0:

    # Basic cleaning: strip column names
    df.columns = df.columns.str.strip()

    # Identify target
    TARGET = 'Price'  # change if different
    if TARGET not in df.columns:
        raise ValueError(f"Target column '{TARGET}' not found in data. Available columns: {df.columns.tolist()}")

    # Quick summary
    print('\n--- Quick stats ---')
    display(df.describe(include='all').T)

    # 4) Train/test split (hold-out test set for final validation)
    train_val, test = train_test_split(df, test_size=0.15, random_state=42)
    train, val = train_test_split(train_val, test_size=0.17647, random_state=42)  # ~15% val so total 70/15/15

    print('\nSplits:', len(train), 'train |', len(val), 'validation |', len(test), 'test')

    # 5) Baseline predictors
    y_train = train[TARGET]
    y_val = val[TARGET]

    baselines = {}
    # Global mean
    baselines['mean'] = {
        'pred': np.repeat(y_train.mean(), len(y_val)),
        'metric': regression_metrics(y_val, np.repeat(y_train.mean(), len(y_val)))
    }
    # Global median
    baselines['median'] = {
        'pred': np.repeat(y_train.median(), len(y_val)),
        'metric': regression_metrics(y_val, np.repeat(y_train.median(), len(y_val)))
    }

    # Group-wise baseline: mean price per number of bedrooms if column exists
    group_col = None
    for candidate in ['number of bedrooms', 'bedrooms', 'beds', 'numbero', 'number o']:
        if candidate in train.columns:
            group_col = candidate
            break

    if group_col is not None:
        grp_mean = train.groupby(group_col)[TARGET].mean()
        # map on val
        preds = val[group_col].map(grp_mean).fillna(y_train.mean())
        baselines[f'group_mean_by_{group_col}'] = {
            'pred': preds.values,
            'metric': regression_metrics(y_val, preds.values)
        }

    print('\nBaseline metrics:')
    for k,v in baselines.items():
        print(k, v['metric'])

    # 6) Feature selection & preprocessing
    # Heuristic: choose numeric columns (excluding id and target) and a few categorical columns
    drop_cols = [c for c in df.columns if 'id' in c.lower()]
    drop_cols += ['Date'] if 'Date' in df.columns else []

    feature_candidates = [c for c in df.columns if c not in drop_cols + [TARGET]]

    # Determine numeric vs categorical
    numeric_cols = df[feature_candidates].select_dtypes(include=['number']).columns.tolist()
    categorical_cols = [c for c in feature_candidates if c not in numeric_cols]

    print('\nNumeric cols\n', numeric_cols)
    print('\nCategorical cols\n', categorical_cols)

    # Create preprocessing pipeline
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])
    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ])

    preprocessor = ColumnTransformer(transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

    # Helper to build full pipeline
    def build_pipeline(model):
        return Pipeline(steps=[('preprocessor', preprocessor), ('regressor', model)])

    # 7) Simple Linear Regression (SLR) with strongest predictor
    # We'll try 'living area' as candidate (common name variations)
    strong_candidates = ['living_area', 'living area', 'living_area ', 'living area sqft', 'livingarea']
    strong_col = None
    for cand in strong_candidates:
        if cand in numeric_cols:
            strong_col = cand
            break

    if strong_col is None and len(numeric_cols) > 0:
        # fallback to the numeric column with highest correlation to target
        corrs = df[numeric_cols + [TARGET]].corr()[TARGET].abs().sort_values(ascending=False)
        # drop target itself
        corrs = corrs.drop(TARGET)
        strong_col = corrs.index[0]
        print(f"Using fallback strong predictor: {strong_col}")
    else:
        print('Using strong predictor:', strong_col)

    # Build SLR using only that predictor
    X_train_slr = train[[strong_col]]
    X_val_slr = val[[strong_col]]

    slr_pipe = Pipeline(steps=[('imputer', SimpleImputer(strategy='median')),
                               ('scaler', StandardScaler()),
                               ('lr', LinearRegression())])
    slr_pipe.fit(X_train_slr, y_train)
    y_val_pred_slr = slr_pipe.predict(X_val_slr)
    print('\nSLR metrics:', regression_metrics(y_val, y_val_pred_slr))
    plot_residuals(y_val, y_val_pred_slr, title=f'SLR residuals ({strong_col})')

    # 8) Multiple Linear Regression (MLR)
    mlr = LinearRegression()
    mlr_pipe = build_pipeline(mlr)
    mlr_pipe.fit(train[feature_candidates], y_train)
    y_val_pred_mlr = mlr_pipe.predict(val[feature_candidates])
    print('\nMLR metrics:', regression_metrics(y_val, y_val_pred_mlr))
    plot_residuals(y_val, y_val_pred_mlr, title='MLR residuals')

    # 9) Check multicollinearity (VIF) - using numeric features only
    if len(numeric_cols) >= 2:
        vif_df = calculate_vif(train[numeric_cols].fillna(train[numeric_cols].median()))
        print('\nVIF table:')
        display(vif_df.sort_values('VIF', ascending=False))

    # 10) Regularization: Ridge, Lasso, ElasticNet with GridSearchCV
    param_grid = {
        'regressor__alpha': [0.01, 0.1, 1.0, 10, 50]
    }

    models = {
        'Ridge': Ridge(random_state=42),
        'Lasso': Lasso(random_state=42, max_iter=5000),
        'ElasticNet': ElasticNet(random_state=42, max_iter=5000)
    }

    model_results = {}
    cv = KFold(n_splits=5, shuffle=True, random_state=42)

    for name, model in models.items():
        pipe = build_pipeline(model)
        gs = GridSearchCV(pipe, param_grid=param_grid, cv=cv, scoring='neg_root_mean_squared_error', n_jobs=-1)
        gs.fit(train[feature_candidates], y_train)
        best = gs.best_estimator_
        # evaluate on validation
        y_val_pred = best.predict(val[feature_candidates])
        metrics = regression_metrics(y_val, y_val_pred)
        model_results[name] = {
            'best_params': gs.best_params_,
            'val_metrics': metrics,
            'estimator': best
        }
        print(f"{name} best params: {gs.best_params_} | val RMSE: {metrics['RMSE']:.2f}")

    # 11) Cross-Validation comparison table (SLR, MLR, Ridge, Lasso, ElasticNet)
    leaderboard = []

    candidates_for_leaderboard = {
        'SLR': slr_pipe,
        'MLR': mlr_pipe,
        'Ridge': model_results['Ridge']['estimator'],
        'Lasso': model_results['Lasso']['estimator'],
        'ElasticNet': model_results['ElasticNet']['estimator']
    }

    for name, estimator in candidates_for_leaderboard.items():
        # Note: SL R expects single column, so handle shapes
        if name == 'SLR':
            X = train[[strong_col]]
        else:
            X = train[feature_candidates]
        # use cross_val_score with neg RMSE via scoring 'neg_root_mean_squared_error'
        scores = cross_val_score(estimator, X, y_train, cv=cv, scoring='neg_root_mean_squared_error', n_jobs=-1)
        mean_rmse = -scores.mean()
        leaderboard.append({'model': name, 'cv_RMSE': mean_rmse})

    leaderboard_df = pd.DataFrame(leaderboard).sort_values('cv_RMSE')
    print('\nLeaderboard (cv RMSE):')
    display(leaderboard_df)

    # 12) Cross-validation strategies demonstration
    # LOOCV (be careful: expensive)
    loo = LeaveOneOut()
    if train.shape[0] <= 200:  # avoid running LOOCV for large datasets
        loocv_score = -cross_val_score(mlr_pipe, train[feature_candidates], y_train, cv=loo, scoring='neg_root_mean_squared_error', n_jobs=-1).mean()
        print('LOOCV RMSE (MLR):', loocv_score)
    else:
        print('Skipping LOOCV due to dataset size')

    # TimeSeriesSplit example (only if you have time/sequence data)
    if 'Date' in df.columns or 'date' in df.columns:
        tscv = TimeSeriesSplit(n_splits=5)
        ts_scores = -cross_val_score(mlr_pipe, train[feature_candidates], y_train, cv=tscv, scoring='neg_root_mean_squared_error', n_jobs=-1)
        print('TimeSeriesSplit RMSE (MLR):', ts_scores.mean())

    # 13) Select champion model
    # We'll choose the model with lowest validation RMSE from our trained regularized models and MLR/SLR
    all_candidates = []
    # add SLR, MLR
    all_candidates.append(('SLR', slr_pipe, regression_metrics(y_val, slr_pipe.predict(val[[strong_col]]))))
    all_candidates.append(('MLR', mlr_pipe, regression_metrics(y_val, mlr_pipe.predict(val[feature_candidates]))))
    for name in model_results:
        all_candidates.append((name, model_results[name]['estimator'], model_results[name]['val_metrics']))

    best = sorted(all_candidates, key=lambda x: x[2]['RMSE'])[0]
    champion_name, champion_model, champion_metrics = best[0], best[1], best[2]
    print(f"\nChampion: {champion_name} with val RMSE: {champion_metrics['RMSE']:.2f}")

    # 14) Final evaluation on hold-out test set
    if champion_name == 'SLR':
        X_test = test[[strong_col]]
    else:
        X_test = test[feature_candidates]
    y_test = test[TARGET]
    test_pred = champion_model.predict(X_test)
    test_metrics = regression_metrics(y_test, test_pred)
    print('Test metrics:', test_metrics)
    plot_residuals(y_test, test_pred, title=f'Champion residuals ({champion_name})')

    # 15) Save artifacts
    out_dir = 'artifacts'
    os.makedirs(out_dir, exist_ok=True)
    joblib.dump(champion_model, os.path.join(out_dir, 'champion_model.joblib'))

    # Save predictions
    test_out = test.copy()
    test_out['predicted_price'] = test_pred
    test_out.to_csv(os.path.join(out_dir, 'houseprice_predictions_test.csv'), index=False)

    # Save metrics
    metrics_all = {
        'baseline': {k: v['metric'] for k,v in baselines.items()},
        'champion': {'name': champion_name, 'val_metrics': champion_metrics, 'test_metrics': test_metrics}
    }
    with open(os.path.join(out_dir, 'test_metrics.json'), 'w') as f:
        json.dump(metrics_all, f, indent=2)

    print('\nSaved artifacts to', out_dir)

    # 16) Short report guidance
    print('\nReport checklist:')
    print('- Document data cleaning decisions (why rows/cols dropped, imputation strategies)')
    print('- Show baseline performance vs models (table + bar chart)')
    print('- Include residual plots and QQ-plots to check normality')
    print('- Provide VIF table and mention any collinearity handling (drop or regularize)')
    print('- Explain why champion model chosen (better CV RMSE, simpler, stable coefficients)')

# End of script
print('\nScript end.')
```
